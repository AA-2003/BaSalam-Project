```python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
import plotly.figure_factory as ff
```

```python
reviews = pd.read_csv('/root/.cache/kagglehub/datasets/radeai/basalam-comments-and-products/versions/1/BaSalam.reviews.csv', low_memory=False)

# reviews = pd.read_csv('data/BaSalam.reviews.csv', low_memory=False)
```

```python
reviews['createdAt'] = pd.to_datetime(reviews['createdAt'])
sells_by_date = reviews['createdAt'].dt.date.value_counts().sort_index()

sells_by_date = pd.DataFrame({
    'day': pd.to_datetime(sells_by_date.index),
    'amount': sells_by_date.values
})
sells_by_date['day_of_week'] = sells_by_date['day'].dt.day_name()

fig = px.line(sells_by_date, x='day', y='amount', title='Daily Sales (2024)',
              labels={'sales': 'Sales Amount', 'day': 'day'},
              hover_data={'day_of_week': True})

fig.update_layout(
    xaxis_title='Date',
    yaxis_title='Sales Amount',
    template='plotly_white',
    hovermode='x unified'
)

fig.show()
```

```python
fig = px.line(sells_by_date[sells_by_date['day'].dt.year >= 2024], x='day', y='amount', title='Daily Sales (2024)',
              labels={'sales': 'Sales Amount', 'day': 'day'},
              hover_data={'day_of_week': True})

fig.update_layout(
    xaxis_title='Date',
    yaxis_title='Sales Amount',
    template='plotly_white',
    hovermode='x unified'
)

fig.show()
```

```python
fig = px.line(sells_by_date[(sells_by_date['day'].dt.year >= 2023) & (sells_by_date['day'].dt.year < 2024)], x='day', y='amount', title='Daily Sales (2023)',
              labels={'sales': 'Sales Amount', 'day': 'day'},
              hover_data={'day_of_week': True})

fig.update_layout(
    xaxis_title='Date',
    yaxis_title='Sales Amount',
    template='plotly_white',
    hovermode='x unified'
)

fig.show()

```
